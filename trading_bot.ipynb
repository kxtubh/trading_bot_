```python
from lumibot.brokers import Apalca
from lumibot.backtesting import YahooDataBacktesting
from lumibot.strategies import Stragtegy
from lumibot.traders import Trader
from datetime import datetime
```

```python
API_KEY=""
API_SECRET=""
BASE_URL=""
```

```python
ALPACA_CREDS=Apalca{"API_KEY":API_KEY,
                    "API_SECRET":API_SECRET,
                    "PAPER":True}

broker=Apalca(APALCA_CREDS)
```

```python
startDate=datetime(2024,1,6)
endDate=datetime(2024,2,6)
```

```python
Class MLTrader(Strategy):
    def initialize(self,symbol:str="SPY"):
      self.symbol=symbol
      self.sleeptime='24H'
      self.last_trade=None

      pass
    def on_trading_iteration(self):
      if self.last_trade == None:
        order=self.create_order(self.symbol,
                                10,
                                'buy',
                                type='market)
                                self.submit_order(order)
                                self.last_trade='buy'

        self.last_trade=self.broker.get_last_trade(self.symbol)
      pass
    broker=Alpaca(AlPACA_CREDS)
    strategy = MLTrader(name='mlstrat',broker=broker,parameters={"symbol":"SPY"})
    strategy.backtest(
        YahooDataBacktesting,
        startDate,
        endDate,
    )
```
